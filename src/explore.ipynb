```python
from sklearn.model_selection import train_test_split

num_variables = ["age", "bmi", "children", "charges"]

x_con_outliers = raw_data_con_outliers.drop("smoker", axis = 1,)[num_variables]
y = raw_data_con_outliers["smoker"]
x_sin_outliers = raw_data_sin_outliers.drop("smoker", axis = 1)[num_variables]
# NO ACABO DE ENTENDER MUY BIEN ESTE PASO

x_train_con_outliers, x_test_con_outliers, y_train, y_test = train_test_split(x_con_outliers, y, test_size = 0.2, random_state = 42)
x_train_sin_outliers, x_test_sin_outliers, _, _ = train_test_split(x_sin_outliers, y, test_size = 0.2, random_state = 42)

y_train.to_csv("../data/processed/y_train.csv", index = False)
y_test.to_csv("../data/processed/y_test.csv", index = False)

x_train_con_outliers.head()
```

NormalizaciÃ³n!!!

```python
from sklearn.preprocessing import StandardScaler
from pickle import dump
import pandas as pd

scaler_con_outliers_norm = StandardScaler()
scaler_con_outliers_norm.fit(x_train_con_outliers)
scaler_sin_outliers_norm = StandardScaler()
scaler_sin_outliers_norm.fit(x_train_sin_outliers)

x_train_con_norm = pd.DataFrame(scaler_con_outliers_norm.transform(x_train_con_outliers), index = x_train_con_outliers.index, columns = num_variables)

x_test_con_norm = pd.DataFrame(scaler_con_outliers_norm.transform(x_test_con_outliers), index = x_test_con_outliers.index, columns = num_variables)

x_train_sin_norm = pd.DataFrame(scaler_sin_outliers_norm.transform(x_train_sin_outliers), index = x_train_sin_outliers.index, columns = num_variables)

x_test_sin_norm = pd.DataFrame(scaler_sin_outliers_norm.transform(x_test_sin_outliers), index = x_test_sin_outliers.index, columns = num_variables)

x_train_sin_norm.head()

dump(scaler_con_outliers_norm, open("../data/processed/scaler_con_outliers_norm.sav", "wb"))
dump(scaler_sin_outliers_norm, open("../data/processed/scaler_sin_outliers_norm.sav", "wb"))
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from pickle import dump

# CON OUTLIERS

scaler_MinMax_con_outliers = MinMaxScaler()
scaler_MinMax_con_outliers.fit(x_train_con_outliers)
scaler_MinMax_sin_outliers = MinMaxScaler()
scaler_MinMax_sin_outliers.fit(x_train_sin_outliers)

x_train_con_scal = scaler_MinMax_con_outliers.transform(x_train_con_outliers)
x_train_con_scal = pd.DataFrame(x_train_con_scal, index = x_train_con_outliers.index, columns = num_variables)

x_test_con_scal = scaler_MinMax_con_outliers.transform(x_test_con_outliers)
x_test_con_scal = pd.DataFrame(x_test_con_scal, index = x_test_con_outliers.index, columns = num_variables)


# SIN OUTLIERS

x_train_sin_scal = scaler_MinMax_sin_outliers.transform(x_train_sin_outliers)
x_train_sin_scal = pd.DataFrame(x_train_sin_scal, index = x_train_sin_outliers.index, columns = num_variables)

x_test_sin_scal = scaler_MinMax_sin_outliers.transform(x_test_sin_outliers)
x_test_sin_scal = pd.DataFrame(x_test_sin_scal, index = x_test_sin_outliers.index, columns = num_variables)

dump(scaler_MinMax_con_outliers, open("../data/processed/scaler_MinMax_con_outliers.sav", "wb"))
dump(scaler_MinMax_sin_outliers, open("../data/processed/scaler_MinMax_sin_outliers.sav", "wb"))

x_train_con_scal.head()
```

```python
from sklearn.feature_selection import f_classif, SelectKBest

# CON OUTLIERS
selection_model_con_outliers = SelectKBest(f_classif, k = 5)
selection_model_con_outliers.fit(x_train_con_outliers, y_train)
column_names = selection_model_con_outliers.get_support()

x_train_sel_con_outliers = pd.DataFrame(selection_model_con_outliers.transform(x_train_con_outliers), columns = x_train_con_outliers.columns.values[column_names])
x_test_sel_con_outliers = pd.DataFrame(selection_model_con_outliers.transform(x_test_con_outliers), columns = x_test_con_outliers.columns.values[column_names])

dump(selection_model_con_outliers, open("../data/processed/selection_model_con_outliers.sav", "wb"))

x_train_sel_con_norm = x_train_con_norm[list(x_train_sel_con_outliers.columns)]
x_train_sel_con_scal = x_train_con_scal[list(x_train_sel_con_outliers.columns)]

x_test_sel_con_norm = x_test_con_norm[list(x_test_sel_con_outliers.columns)]
x_test_sel_con_scal = x_test_con_scal[list(x_test_sel_con_outliers.columns)]


# SIN OUTLIERS
selection_model_sin_outliers = SelectKBest(f_classif, k = 5)
selection_model_sin_outliers.fit(x_train_sin_outliers, y_train)
column_names = selection_model_sin_outliers.get_support()

x_train_sel_sin_outliers = pd.DataFrame(selection_model_sin_outliers.transform(x_train_sin_outliers), columns = x_train_sin_outliers.columns.values[column_names])
x_test_sel_sin_outliers = pd.DataFrame(selection_model_sin_outliers.transform(x_test_sin_outliers), columns = x_test_sin_outliers.columns.values[column_names])

dump(selection_model_sin_outliers, open("../data/processed/selection_model_sin_outliers.sav", "wb"))

x_train_sel_sin_norm = x_train_con_norm[x_train_sel_sin_outliers.columns]
x_train_sel_sin_scal = x_train_sin_scal[x_train_sel_sin_outliers.columns]

x_test_sel_sin_norm = x_test_sin_norm[x_test_sel_sin_outliers.columns]
x_test_sel_sin_scal = x_test_sin_scal[x_test_sel_sin_outliers.columns]
```

Machine Learning (Linear Regression)

```python
train, test = [], []

for name, df in dfs_train.items():
    train.append(df)

for name, df in dfs_test.items():
    test.append(df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import warnings
from sklearn.metrics import root_mean_squared_error

results = []

for index in range(len(train)):
    model = LinearRegression()
    train_df = train[index]
    model.fit(train_df, y_train)
    y_test_pred = model.predict(test[index])

    results.append(
        {
            "index" : index,
            "df_train" : list(dfs_train.keys())[index],
            "Coeficiente" : model.coef_,
            "MAE" : mean_absolute_error(y_test, y_test_pred).round(2),
            "RMSE" : round(root_mean_squared_error(y_test, y_test_pred), 2),
            "R2_score" : round(r2_score(y_test, y_test_pred), 5)
        }
    )

results = sorted(results, key = lambda x: x["RMSE"])
best_ind = results[0]["index"]
best_df_train = results[0]["df_train"]

results
```
